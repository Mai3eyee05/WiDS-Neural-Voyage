```python
arr2 = np.array([1,2,3])
arr = np.arange(1,25,2).reshape(3,4)
print("arr2:\n", arr2)
print("arr:\n", arr)
result = np.multiply(arr2,arr.T)
print(result)
result2 = np.matmul(arr2,arr)
print(result2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('iris_dataset.csv')
index = np.arange(150)
feature_1= 'sepal length (cm)'
feature_2= 'sepal width (cm)'
feature_3= 'petal width (cm)'
feature_4= 'petal length (cm)'
feature_5= 'target (cm)'


feature_1_data = df['sepal length (cm)'].iloc[:150].values
feature_2_data = df['sepal width (cm)'].iloc[:150].values
feature_3_data = df['petal width (cm)'].iloc[:150].values
feature_4_data = df['petal length (cm)'].iloc[:150].values
feature_5_data = df['target'].iloc[:150].values


plt.plot(index, feature_1_data, label='sepal length', color='g', linestyle='-', marker='x')
plt.plot(index, feature_2_data, label='sepal width',color='r', linestyle='--', marker='s')
plt.plot(index, feature_3_data, label='petal length', color='b', linestyle='-', marker='o')
plt.plot(index, feature_4_data, label='petal width',color='y', linestyle='--', marker='x')
plt.plot(index, feature_5_data, color='g', linestyle='-', marker='s')

plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.title('Line Plot for Each Feature')
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, loc='upper right', bbox_to_anchor=(1,1))

plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('iris_dataset.csv')
target_map= {0: ('Setosa', 'red'), 1: ('Versicolor', 'blue'), 2:('Virginica', 'green')}
df['species_name'] = df['target'].map({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})
for target_value, (species_name, color) in target_map.items():
    species_data = df[df['species_name'] == species_name]
    plt.scatter(species_data['petal length (cm)'], species_data['petal width (cm)'],
                label=species_name, color=color)



plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend(loc='upper right', bbox_to_anchor=(1,1))
plt.tight_layout()

plt.show()
print(df['target'].unique())
```
